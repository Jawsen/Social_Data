### **What are the three key elements to keep in mind when you design an explanatory visualization?**

The three key elements to keep in mind when you design a explanatory visualization is the following:
1) Start with a question 
2) Allow exploration 
3) Know your readers

### Explain in your own words: How is explanatory data analysis different from exploratory data analysis?

- Exploratory data analysis (EDA) is the process of exploring data to identify patterns, trends, and possible relationships. It is the initial phase where unknowns are investigated, questions are formed, and insights emerge through analysis and visualization.

- Explanatory data analysis, on the other hand, occurs after the data has been explored and tested. At this stage, findings are structured and presented clearly to communicate insights effectively to an audience. The focus shifts from discovery to explanation.
